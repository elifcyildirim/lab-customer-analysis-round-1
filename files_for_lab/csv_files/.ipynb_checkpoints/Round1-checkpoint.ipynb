```python
file1 = file1.rename(columns={'ST':'State'})
file1.columns
```

```python
file2 = file2.rename(columns={'ST':'State'})
file2.columns
```

```python
cols = []
for i in range(len(file1.columns)):
    cols.append(file1.columns[i].lower().replace(' ','_'))
file1.columns = cols
file1.head()
```

```python
cols = []
for i in range(len(file2.columns)):
    cols.append(file2.columns[i].lower().replace(' ','_'))
file2.columns = cols
file2.head()
```

```python
cols = []
for i in range(len(file3.columns)):
    cols.append(file3.columns[i].lower().replace(' ','_'))
file3.columns = cols
file3
```

```python
# 4 Rearrange the columns in the dataframe as needed
```

```python
file1.columns
```

```python
# # Rearranging columns: how to #also dropping some of the columns. I only write the names of the columns that I am interested in.

file2 = file2[['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
file2
```

```python
# # Rearranging columns: how to #also dropping some of the columns. I only write the names of the columns that I am interested in.

file3 = file3[['customer', 'state', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
file3
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so divide them by 100 and change dtype to numerical type.

file1['customer_lifetime_value'] = file1['customer_lifetime_value'].str.replace('%','')
file1['customer_lifetime_value'] = pd.to_numeric(file1['customer_lifetime_value'], errors='coerce')
file1['customer_lifetime_value'] = file1['customer_lifetime_value']/100

file2['customer_lifetime_value'] = file2['customer_lifetime_value'].str.replace('%','')
file2['customer_lifetime_value'] = pd.to_numeric(file2['customer_lifetime_value'], errors='coerce')
file2['customer_lifetime_value'] = file2['customer_lifetime_value']/100
```

```python
# Concatenate the three dataframes
```

```python
data = pd.concat([file1, file2, file3], axis=0) #axis=0 means columns & axis=1 means rows
data.shape
```

```python
data
```

```python
data=data.reset_index(drop=True) #starts the row index from 0
data
```

```python
## 5 Which columns are numerical?

    #ANSWER: These columns are numerical: income, monthly premium auto, total claim amount

# 6 Which columns are categorical?

    ##ANSWER: All columns except income, monthly premium auto, total claim amount are strings and therefore categorical.

# 7 Understand the meaning of all columns

data.dtypes
```

```python
#Delete the column education and the number of open complaints from the dataframe.

data = data.drop(['education','number_of_open_complaints'], axis=1) #axis=1 means rows
```
