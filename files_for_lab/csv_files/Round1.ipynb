```python
# # Rearranging columns: how to #also dropping some of the columns. I only write the names of the columns that I am interested in.
file2 = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file2
```

```python
# # Rearranging columns: how to #also dropping some of the columns. I only write the names of the columns that I am interested in.
file3 = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file3
```

```python
# Concatenate the three dataframes
```

```python
data = pd.concat([file1, file2, file3], axis=0) #axis=0 means columns & axis=1 means rows
data.shape
```

```python
#Delete the column education and the number of open complaints from the dataframe.

data = data.drop(['education','number of open complaints'], axis=1) #axis=1 means rows
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so divide them by 100 and change dtype to numerical type.
for i in range(len(data)):
    if ('%' in data['customer lifetime value']):
        data['customer lifetime value'] = data['customer lifetime value'].str.rstrip('%').astype('float') / 100.0

#data['customer lifetime value'] =  pd.to_numeric(data['customer lifetime value'], errors='coerce')
```
